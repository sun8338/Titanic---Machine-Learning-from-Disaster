##Preprocessing

Encode Sex

```python
train["Sex_encode"]=train["Sex"].replace("male",0).replace("female",1)
```

```python
print(train.shape)
```

```python
train[["Sex","Sex_encode"]].head()
```

```python
test["Sex_encode"]=test["Sex"].replace("male",0).replace("female",1)
```

```python
test.shape
```

```python
test[["Sex","Sex_encode"]].head()
```

Fill in missing Fare

```python
test["Fare"]=test["Fare"].fillna(0)  
test[test["Fare"].isnull()] 
```

##Encode Embarked

```python
# C==O, S==1, Q==2


train["Embarked_C"]=train["Embarked"]=="C"
train["Embarked_S"]=train["Embarked"]=="S"
train["Embarked_Q"]=train["Embarked"]=="Q"
print(train.shape)
train
```

```python
train[["Embarked_C","Embarked_S","Embarked_Q"]].head()
```

```python
test["Embarked_C"]=test["Embarked"]=="C"
test["Embarked_S"]=test["Embarked"]=="S"
test["Embarked_Q"]=test["Embarked"]=="Q"
print(test.shape)
test
```

```python
test[["Embarked_C","Embarked_S","Embarked_Q"]].head()
```

##Age를 featuring으로 사용하기

```python
train["Age"].isnull() 
```

```python
n_mean=np.mean(train["Age"])
n_mean
```

```python
train["Age"]=train["Age"].fillna(n_mean)  
```

```python
train["Age"].head(6)
```

```python
train["Age_X"]=train["Age"]<15
train["Age_X"].head(17)
```

```python
test["Age"].isnull() 
```

```python
n_mean=np.mean(test["Age"])
n_mean
```

```python
test["Age"]=test["Age"].fillna(n_mean) 
```

```python
test["Age"].head(20)
```

```python
test["Age_X"]=test["Age"]<15
test["Age_X"].head(17)
```

Train

```python
#Feature-Pclass, Sex, Fare, Age
#Label-Survived
```

```python
feature_names=["Pclass","Sex_encode","Fare","Embarked_C","Embarked_S","Embarked_Q","Age_X"]
```

```python
feature_names
```

```python
x_train=train[feature_names]
```

```python
print(x_train.shape)
```

```python
x_train.head()
```

```python
x_test=test[feature_names]
```

```python
print(x_test.shape)
```

```python
x_test.head()
```

```python
label_name="Survived"
```

```python
y_train=train[label_name]
```

```python
print(y_train.shape)
```

```python
y_train.head() #y_test가 없는 것은 우리가 맞쳐야 하는 값이고 실제로 y_test의 값은 없음. 
```

Use Decision Tree

```python
#scikit-learn==sklearn
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(max_depth=15)
```

```python
#1)fit(train)/2)predict(test) fit은 머신러닝을 학습시키는것. 패턴을 스스로 판단하는 행위. 
model.fit(x_train, y_train)
```

```python
prediction=model.predict(x_test)
```

```python
print(prediction.shape)
prediction[0:5]
```
